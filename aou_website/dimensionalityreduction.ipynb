### PCA


Dimensionality Reduction with PCA:
Apply PCA to your record dataset.
Determine the optimal number of principal components to retain.
Visualize the reduced-dimensional data using PCA.
Analyze and interpret the results.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ####GET DATASET
# from keras.datasets import mnist
# (X, Y), (test_images, test_labels) = mnist.load_data()


from sklearn.datasets import load_digits
# LOAD DATA
X= newAppledf
Y= y

print(X.shape)

#NORMALIZE AND RESHAPE
X=X/np.max(X) #NORMALIZE
# X=X.reshape(60000,28*28); #print(X[0])

#COMPUTE PCA
from sklearn.decomposition import PCA
n_components=5
pca = PCA(n_components=n_components)
# pca.fit(X)
X1=pca.fit_transform(X)

#2D PLOT
plt.scatter(X1[:,1], X1[:,2], c=Y, cmap='tab10')
plt.show()
```

### t- SNE at Different Perplexity Values

```python
from sklearn.datasets import load_digits
import numpy as np
from sklearn.manifold import TSNE
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt

# LOAD DATA
print("LOAD MNIST DATA")
X= newAppledf
print(X.shape)

# DO DIMENSIONALITY REDUCTION
X_embedded = TSNE(n_components=2, learning_rate='auto',init='random', perplexity=7).fit_transform(X)

# EXPLORE RESULTS
print("RESULTS") 
print("shape : ",X_embedded.shape)
print("First few points : \n",X_embedded[0:2,:])

# PLOT 
plt.scatter(X_embedded[:,0],X_embedded[:,1], alpha=0.5)
plt.show()
```
